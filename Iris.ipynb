# Iris DataSet


```python
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
```

```python
data = pd.read_csv('iris.csv')
```

```python
data.drop('Id', axis=1, inplace=True)
```

```python
data
```

```python
import warnings as wa
wa.filterwarnings("ignore")
```

```python
sns.pairplot(data=data, hue='Species',palette='Paired')
```

```python
le = preprocessing.LabelEncoder()
le.fit(data.Species)
data['ClassOfSpecies'] = le.transform(data.Species)
```

```python
data
```

```python
X = data.iloc[:,:4]
Y = data.iloc[:,5:6]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.3)
```

```python
model = SVC(gamma='scale')

model.fit(X_train,y_train)
```

```python
predication = model.predict(X_test)
```

```python
print("Accuracy : ",metrics.accuracy_score(y_test,predication))
```
